```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
file = pd.read_csv("https://raw.githubusercontent.com/xeqoir/Finance/refs/heads/main/FamaFrenchThreeFactorsMonthly.csv")
file = file.apply(pd.to_numeric, errors='coerce').dropna()
file
```

```python
# Separate factors and dependent variable
X_3factor = file[['Mkt-RF', 'SMB', 'HML']]  # 3-factor model inputs
y = file['Mkt-RF']  # Dependent variable (e.g., asset return - risk-free rate)
X_3factor = sm.add_constant(X_3factor)  # Add intercept
```

```python
model_3factor = sm.OLS(y, X_3factor).fit()
print("Fama-French 3-Factor Model Results:")
print(model_3factor.summary())
```
